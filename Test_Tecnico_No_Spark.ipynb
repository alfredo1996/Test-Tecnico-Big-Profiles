```python
"""
Given a list of numbers, returns its cumulative sum
"""
def windows_cum_sum(values):
  output = list()
  total = 0
  for value in values:
      total += value
      output.append(total)
  return output

"""
Given a list of values, return a list containing a partition
of the input in n lists
"""
def partition(lst, n):
  division = len(lst) / n
  return [lst[round(division * i):round(division * (i + 1))] for i in range(n)]


"""
Given a list of numbers, return its cumulative sum 
for each window of len(list)/number_of_partitions numbers
"""
def cumulative_sum(lines, number_of_partitions = 100):

  windows = partition(lines_int,number_of_partitions)

  sums = [sum(window) for window in windows]

  #Returns the input's cumulative sum
  cumulative_sums = windows_cum_sum(sums)

  #Store the values in a dictionary to call them using an index between [1,100]
  dict_sums = dict()
  index = 1

  for value in cumulative_sums:
    dict_sums.update({index:value})
    index = index + 1

  return dict_sums
```

```python
from collections import defaultdict
import operator

#Initialize the dictionary
dictionary = defaultdict(list)

with open('/content/sample_data/exercise_2.txt') as f:
    lines = f.read().splitlines()
    for line in lines[1:]: #We remove the first line because it is the header
      row = line.split(",")
      dictionary[row[0]].append(float(row[1]))

#Initialize total and count on which we will find the mean later
total = 0.0
count = 0.0
aggregated_means = defaultdict(float) 
for key in dictionary.keys():
  key_sum = sum(dictionary[key])
  key_count = len(dictionary[key])
  aggregated_means[key] = key_sum/key_count
  total += key_sum
  count += key_count

#Find the total mean
total_mean = total/count

#Find the distances, for each group, from the 
distances = {}
for key in dictionary.keys():
  distances.update({key : total_mean - aggregated_means[key]})

#Sort the keys on the values of the dictionary 
sorted_tuples = sorted(distances.items(), key=operator.itemgetter(1), reverse = True)
sorted_distances = {k: v for k, v in sorted_tuples}

for key in sorted_distances.keys():
  print("Key: "+ key +" Value: "+ str(sorted_distances[key]) +"")
```
